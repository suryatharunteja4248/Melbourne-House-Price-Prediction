```python
#FeatureEngineering
x['Date']=pd.to_datetime(x['Date'],errors='coerce',dayfirst= True)
x['Year']=x['Date'].dt.year
x['Month']=x['Date'].dt.month
x['Day']=x['Date'].dt.day

x.drop(columns = ['Date'], axis = 1, inplace = True)
```

```python
num_x = x.select_dtypes(include="number")
cat_x = x.select_dtypes(exclude="number")
```

```python
# preprocessing

num_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaling", StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_transformer, num_x),
        ("cat", cat_transformer, cat_x)
    ],
    remainder="drop"
)

preprocessor
```
